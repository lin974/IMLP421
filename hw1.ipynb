### 02 - 填字遊戲（Mad libs game）
#### 使用者輸入任何字，它可以是名詞、形容詞、動詞、代名詞等。程式得到文字後，可自行排列，組成成一個模板故事。

> 題目：輸入一數字判斷奇數或偶數<br>
輸入(Input): 一連串依據說明輸入對應文字 <br>
輸出(Output): 模板故事

* 進階: 可以讓程式重複執行

````python
a = int(input("請需要建立幾次模板故事"))
for b in range(a):
    num = int(input("請輸入課程期數"))
    class_name = input("請輸入課程名稱")
    name = input("請輸入姓名")
    background = input("請輸入你的背景(科系或職業)")
    hope = input("請簡述對於本課程的期望(列出幾點修完課程想得到的技能/對自己未來的規劃)")

    print(f"第{num}期-{class_name}")
    print(f"姓名:{name}")
    print(f"學生背景: {background}")
    print("==================================")
    print(f"{name} 修習完課程[{class_name}]後，")
    print(f"結業獲得{hope}。")
    print("well done")

````

### 03 -  猜數字（Guess the number）
#### 請使用者猜一個介於範圍（例如 1 到 100）之間的數字，
#### 若使用者猜錯，就詢問他們想繼續玩還是退出；若使用者猜對，就顯示祝賀訊息，
#### 並統計使用者的嘗試次數。如果使用者輸入的數字超出設定範圍，就顯示錯誤提示。

![](data\guess.png)


````python
import random 
ans = random.randint(1,100)
num = int(input("你猜的數字"))
low = 0
high = 100
while num != ans: 
    if num>100 or num <0 or num>high or num<low:
        print("error")
        num  = int(input(f"在猜一遍,介於{low}-{high}"))
    elif num>ans:
        high = num
        num  = int(input(f"在猜一遍,介於{low}-{high}"))
       
    elif num<ans:
        low = num
        num  = int(input(f"在猜一遍,介於{low}-{high}"))
   
print("恭喜猜中")
````

### 05 - Email 域名判斷器（Email slicer）
#### 請用戶輸入 Email 地址，然後判斷它是自定義域名還是熱門域名。

> 題目：<br>
輸入(Input):shelly200318@hotmail.com.tw  <br>
輸出(Output):Your username is 'shelly200318' and your domain name is 'hotmail.com.tw' <br>


* 進階：把常用的信箱存成字典，加入判斷是否為
>輸入(Input):mary.jane@gmail.com  <br>
輸出(Output):這是註冊在 Google 之下的 Email 地址 <br>
輸入(Input):matt.pan@myfantasy.com  <br>
輸出(Output):這是在 myfantasy 之下自定義域



````python
email = input("what is your email adress").strip()
position = email.rfind("@")
num = len(email)
user_pos = position - 1
domain_pos = position + 1
user_name = " "
domain_name = " "
for a in range(user_pos+1):
    user_name += email[a]
for b in range(domain_pos,num):
    domain_name += email[b]
print("your user name is ",user_name)
print("your domain name is",domain_name)
````

### 06 - 及格名單（Pass list）
#### 請使用集合功能來完成各科級名單的判斷
```
米花市帝丹小學一年级B班正舉辦期中考試
數學及格的有：柯南、灰原、步美、美環、光彦
英文及格的有：柯南、灰原、丸尾、野口、步美
以上已列出全班所有人
```

>請分別列出<br>
數學及格但英文不及格的同學名單<br>
數學不及格但英文及格的同學名單<br>
兩者皆及格名單<br>
* Hint:差集(減法)、交集


````python
m = {'柯南','灰原','步美','美環','光彦'}
e = {'柯南','灰原','丸尾','野口','步美'}
a = m.intersection(e)
b = m - a
c = e - a
print(f"數學及格但英文不及格 {b}")
print(f"數學不及格但英文及格 {c}")
print(f"兩者皆及格 {a}")
````

### 09 - 回文判斷（Is a palindrome）
#### 請使用者輸入單字，判斷它是否為回文，也就是該單字前後對稱，例如 madam，從前讀到後或是從後讀到前的順序都是 madam。

> 題目：<br>
輸入(Input): 雨 滋 春 樹 碧 連 天  天 連 碧 樹 春 滋 雨 <br>
輸出(Output):The text you have entered is a palindrome!<br>
輸入(Input): 資工訓練班 <br>
輸出(Output):The text you have entered is not a palindrome.


````python
a = input()
len_a = len(a)
rev_a = []


for b in range(len_a-1,-1,-1):
    
    rev_a += a[b]



lista = list(a)

rev_list = list(rev_a)


if rev_list == lista:
    print("it is palindrome")
else:
    print("not paildrome")
````
